```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
brightness_treshold = [77,76,79,83,86,88,83,89,72,73,74,75,75,73,74]
positions = [15, 15.2, 15.4, 15.6, 15.8, 16, 14.8, 14.6, 15.1, 15.12, 15.08, 15.14, 15.06, 15.16, 15.18]
data1 = np.array([positions, brightness_treshold])
```

```python
fs=17
plt.figure(figsize=(8,6))
plt.plot(data1[0]-15.11, data1[1], "o")
plt.plot([0, 0],[min(data1[1]), max(data1[1])], c="gray", linestyle="--")
plt.xlabel("relative lents position (mm)", fontsize=fs)
plt.ylabel("brightness treshold (a.u.)", fontsize=fs)
plt.show()
```
