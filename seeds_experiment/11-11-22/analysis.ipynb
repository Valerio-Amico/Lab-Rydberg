```python
data = open("data.txt", "r")
data_all = []
for row in data:
    row = row.replace("\n", "")
    row_list_aus = row.split("\t")
    row_list = [float(row_list_element) for row_list_element in row_list_aus] 
    data_all.append(row_list)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
df = pd.DataFrame(data_all, columns=["alignment", "with", "with_std", "seeds", "seeds_std", "without", "without_std", "diff", "diff_std", "power_IR", "N_atoms", "peak_density", "radius_MOT", "IR_std", "time", "seed_time"])
df
```

```python
df.corr(min_periods=3)
```
