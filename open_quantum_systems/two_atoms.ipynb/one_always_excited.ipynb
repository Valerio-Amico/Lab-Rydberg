```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as scc
```

```python
def V(r): # r is in m
    C6 = 869.7 * scc.giga * scc.micron**6
    return C6/r**6
```

```python
detunings = [0] + list(np.logspace(0,3,10))

plt.figure(figsize=(12,6))

plt.subplot(1,2,2)
for i, detuning in enumerate(detunings):
    X = np.linspace(0, 1.7e1, 100000)*1e-5
    Y = 1/(1+(V(X)-detuning)**2)*(1+(-detuning)**2)

    Y1 = len(X)*[1/(1+(-detuning)**2)]

    plt.plot(X*1e5,Y, label="%d MHz"%detuning, color="blue", alpha=1-i/len(detunings))
    #plt.plot(X*1e5,Y1, color="gray", linestyle="--", alpha=0.3)
    
plt.plot(X*1e5,len(X)*[1], color="gray", linestyle="--", alpha=0.3)
plt.xlabel(r"interatomic distance [$\mu m$]",fontsize=16)
plt.ylabel("excitation rate [a.u.]", fontsize=16)
#plt.legend(fontsize=14)
plt.ylim([1e-2,1e5])
plt.yscale("log")

plt.subplot(1,2,1)
for i, detuning in enumerate(detunings):
    X = np.linspace(0, 1.7e1, 100000)*1e-5
    Y = 1/(1+(V(X)-detuning)**2)

    Y1 = len(X)*[1/(1+(-detuning)**2)]

    plt.plot(X*1e5,Y, label="%d MHz"%detuning, color="blue", alpha=1-i/len(detunings))
    plt.plot(X*1e5,Y1, color="gray", linestyle="--", alpha=0.3)

plt.xlabel(r"interatomic distance [$\mu m$]",fontsize=16)
plt.ylabel("excitation rate [a.u.]", fontsize=16)
#plt.legend(fontsize=14)
plt.ylim([0,1.2])

#plt.savefig('olways_one_exc.png', dpi=300)
plt.show()
```
