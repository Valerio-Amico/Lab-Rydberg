```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

$\alpha$ is the probability to have no seed.

```python
def N(alpha, N1, N2):
    if np.random.random() < alpha:
        return N1
    else:
        return np.random.poisson(N2)
    
al = 0.2
N_1 = 0
N_2 = 6

n_sample = 10000
Ns = [N(al, N_1, N_2) for _ in range(n_sample)]

mean = np.mean(Ns)
std = np.std(Ns)

Q = std**2/mean - 1

print("Q = ", Q)
bins = np.arange(0,20,1)
plt.hist(Ns, bins=bins, density=True)#, weights=len(Ns)*[1/sum(Ns)])

N_s_p = poisson.pmf(bins, mean)

plt.plot(bins, N_s_p, color="black", ls="--", marker=".", alpha=0.5)

plt.show()
```
