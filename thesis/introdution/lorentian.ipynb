```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lor(x, delta, w):
    return 1/(w+(delta-x)**2)

def gaus(x, x0, s):
    return 1/np.sqrt(2*np.pi*s**2) * np.e**(-(x-x0)**2/(2*s**2))

xs = np.linspace(-15,15,1000)
ys = lor(xs, 0, 10)

x_g = np.linspace(2,10,1000)
y_g = gaus(x_g, 6, 1)/10 + 0.02

plt.plot(xs, ys)
plt.plot(x_g, y_g)
plt.show()
```
