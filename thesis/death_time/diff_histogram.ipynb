```python
t_range = 4.4 # how many ions arrived in one experiment
reps = 100000
death_time = 6 # bethween 0 and 1
diffs3 = []
range_ = 40

for _ in range(reps):

    t_range_ = np.random.poisson(t_range)
    t_i = np.sort(np.random.random(t_range_)*range_)
    if len(t_i)>0:
        t_lim = t_i[0]
        for j in range(t_range_):
            if t_i[j] - t_lim > death_time+(np.random.normal()*range_)/300: # random death time
                diffs3 = diffs3 + [t_i[j] - t_lim]
                t_lim = t_i[j]

X = np.linspace(0,range_,1000)
Y = np.array([p_delta_t(x, t_range, death_time) for x in X])/1
plt.plot(X,Y)
plt.hist(diffs3, bins=np.linspace(0,range_,40), density=True)
plt.plot([6, 6], [0,0.1], color="gray")
plt.show()
```
