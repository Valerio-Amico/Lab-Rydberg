the probability is
$$
p(\Delta t| N; \tau) = N e^{-N(\Delta t -\tau)} \ \ \ \ \ \  (\Delta t >\tau)
$$
$$
p(\Delta t; \tau) = \sum_N^{\infty} p(\Delta t| N; \tau) p(N) = \sum_N^{\infty} N e^{-N(\Delta t -\tau)} p(N)
$$
if we assume for $\tau$ a gaussian distribution
$$
p(\Delta t| N, \hat{\tau}) = \int_0^{\infty}  p(\Delta t| N, \tau) p(\tau| \hat{\tau}) d\tau
$$
with 
$$
p(\tau| \hat{\tau}) = \frac{1}{\sqrt{2 \pi \sigma_{\tau}^2}} e^{-\frac{(\tau-\hat{\tau})^2}{2\sigma_{\tau}^2}}
$$
so the integral is
$$
p(\Delta t| N, \hat{\tau}) = \int_0^{\infty} N e^{-N(\Delta t -\tau)} \frac{1}{\sqrt{2 \pi \sigma_{\tau}^2}} e^{-\frac{(\tau-\hat{\tau})^2}{2\sigma_{\tau}^2}} d \tau
$$
