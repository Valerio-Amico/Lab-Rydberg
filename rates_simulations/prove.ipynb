```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
equations = np.array([
    [1, -1],
    [0, 0]
])

results = {"1": [], "2": []}

F = lambda t, s: np.dot(equations, s)

t_eval = np.arange(0, 1, 0.01)
sol = solve_ivp(F, [0, 1], [0, 10], t_eval=t_eval)

results["1"].append(sol.y.T[:, 0])
results["2"].append(sol.y.T[:, 1])

plt.figure(figsize = (18, 10))

for j, level_plotted in enumerate(results.keys()):
    
    k = 121+j
    plt.subplot(k)
    plt.plot(t_eval, results[level_plotted][0])

    plt.xlabel('t')
    plt.ylabel(level_plotted)
    plt.legend()
    
    plt.grid()

print(results["1"][0][-1]+results["2"][0][-1])
```
