```python
t=np.linspace(0,40,1000)
beta=0.1
for A in [0.78, 2.12, 3.52, 2.25, 4.45, 9.04]:
    N=A/beta*(1-np.exp(-beta*t))
    plt.plot(t, N, label='Exc rate=%.2f MHz'%A)

plt.xlabel('Time $[\mu s]$')
plt.ylabel('N')
plt.legend()
plt.grid()
plt.text(22,60,'Depumping rate = 0.1 MHz')
plt.savefig('Model.jpg', dpi=1200)
plt.show()
```

Rate as a function of the power 
<br >23.397 is the angular coefficient of the  resonant excitation rate

```python
P30=np.linspace(0,30,100)
R30=P30*23.397/(2*pi)
plt.xlabel('P [mW]')
plt.ylabel('Excitation Rate [kHz]')
plt.plot(P30, R30)

```

Here I correct the excitation rate including the detuning (factor $\frac{\Delta^2}{\gamma^2}$)
<br>Excitation Rate with the detuning
<br>The interested zone is near 157 MHz

```python
P30=30
Delta=np.linspace(0,180,180)
Gamma70=2*np.pi*0.7
Gamma30=(P30*23.397)*(1/(1+(2*np.pi*Delta/Gamma70)**2))
plt.xlabel('Detuning from 70S[MHz]')
plt.ylabel('Off-resonant excitation rate [kHz]')
plt.xlim(80,170)
plt.ylim(0,0.1)
plt.plot(Delta, Gamma30)
plt.savefig('Off-resonant rate.jpg', dpi=1200)
plt.show()
print(Gamma30[157])

```
