```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def delta_blue(delta_f):
    return 164.82 - delta_f
```

```python
delta_fs = np.linspace(0,80,1000)
delta_blues = delta_blue(delta_fs)

plt.plot(delta_fs, delta_blues)
plt.grid()
plt.ylabel("$\Delta$_blue [MHz]", fontsize=16)
plt.xlabel("$\Delta$_f [MHz]", fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
